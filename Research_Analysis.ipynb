## 3. Why MindMap is Best? (Core Techniques)

Our project achieves 'State-of-the-Art' performance using these unique techniques:

### 1️⃣ Multi-Modal Feature Engineering
Unlike traditional surveys that only ask "Are you stressed?", we capture **Lifestyle Proxies**:
- **Sleep Racing:** Identifying racing thoughts before sleep as a proxy for anxiety.
- **Friday Mood:** Evaluating cumulative weekly exhaustion.
- **Commute/Living State:** Understanding the environmental impact (Hostel vs Home).

### 2️⃣ Hybrid Unleashed Modeling
- **XGBoost (Extreme Gradient Boosting):** Handles complex, non-linear relationships between social media usage and academic pressure far better than standard SVM.
- **K-Means Clustering:** Instead of just a binary "Yes/No" stress, we provide **Psychological Archetypes** (e.g., 'The Overloaded Achiever'), which provides more humanized feedback.

### 3️⃣ Local & Regional Optimization
- Tailored for the **Pune Student Ecosystem** (COEP, VIT, MIT links).
- **Multilingual Support (Marathi/Hindi):** Ensures data collection is bias-free for students more comfortable in regional languages, a factor often ignored in global research papers.

### 4️⃣ Instant Actionability
- Traditional research stops at the "Score". MindMap generates a **dynamic 5-page PDF report** with personalized book/game recommendations and daily routines.
